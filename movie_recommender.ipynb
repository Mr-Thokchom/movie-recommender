```python
import numpy as np 
import pandas as pd 
```

```python
movies= pd.read_csv("tmdb_5000_movies.csv")
credits= pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies= movies.merge(credits, on="title")
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
# movie_id
# cast
#crew
#genres
#overview
#title
#keywords
movies= movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"
#Lets convert this into ["Action","Adventure","Fantasy","Scifi"]
```

```python
import ast
def convert(obj):
    List=[]
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
#Lets take up only the first 4 main actors in each movies
def convert4(obj):
    List=[]
    number=0
    for i in ast.literal_eval(obj):
        if number !=4:
            List.append(i['name'])
            number+=1
        else:
            break
    return List
```

```python
movies['cast']=movies['cast'].apply(convert4)
```

```python
movies.head(1)
```

```python
#We will try to take only the director name
def fetch_director(obj):
    List=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            List.append(i['name'])
            break
    return List
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#The column overview is a string so lets convert it into a list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#removing the spaces
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(2)
```

```python
df= movies[['movie_id','title','tags']]
```

```python
df.head(2)
```

```python
#Lets convert the tag into strings
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
#lets converts tags into lower case
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors= cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
def recommend(movie):
    #getting index no. of particular movie
    movie_index =df[df['title']== movie].index[0]
    #calculating similarity distance of that given movie w.r.t index
    distance=similarity[movie_index]
    #sorting most similiar first 5 movies w.r.t given movie 
    movie_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:6]
    #printing those 5 movies index no.
    for i in movie_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
